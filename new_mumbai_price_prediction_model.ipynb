```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Mumbai.csv')
df.head()
print(df.shape)
```

```python
df2 = df.copy()
df2.columns
```

```python
df2 = df2.drop_duplicates()
df2.shape
```

```python
df3 = pd.get_dummies(df2).astype(int)
df3.head()
df3.columns
```

```python
df3.corr()
```

```python
for i in df3.columns:
  print(df3[['Price',i]].corr())
```
